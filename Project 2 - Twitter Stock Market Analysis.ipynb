Task 2: Statistical Tests and Analysis (Optional)

1.Give me the Z-test or T-test over High, Low, and
Close columns and see if the null hypothesis gets
rejected or accepted:

```python
import statistics
from scipy import stats
def ttest (col1, col2, Twitter):
    t_stat, p_value = stats.ttest_rel(Twitter[col1], Twitter[col2])
    result = "Reject Null Hypothesis (significant difference)" if p_value < 0.05 else "Fail to Reject Null Hypothesis"
    return {"T-statistic": t_stat, "P-value": p_value, "Result": result}

test_high_low = ttest("High", "Low", Twitter)
test_high_close = ttest("High", "Close", Twitter)
test_low_close = ttest("Low", "Close", Twitter)

print(test_high_low)
print(test_high_close)
print(test_low_close)
```

Task 3: Visualization and Analysis

1. Show the Twitter stock prices over the years and
give a conclusion:

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Group by Year and take average Open price
yearly_open = Twitter.groupby('Year')['Open'].mean()

plt.figure(figsize=(12,6))
plt.plot(yearly_open.index, yearly_open.values, marker='o', label='Avg Open Price')

plt.show()
```

3. Assign buttons to control time periods. Add the
buttons to analyze the stock prices of Twitter in
different time periods:

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(x=Twitter['Date'], y=Twitter['Close'],
                         mode='lines', name='Close Price'))

fig.update_layout(
    title="Twitter Stock Prices Over Time",
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1, label="1M", step="month", stepmode="backward"),
                dict(count=6, label="6M", step="month", stepmode="backward"),
                dict(count=1, label="1Y", step="year", stepmode="backward"),
                dict(count=5, label="5Y", step="year", stepmode="backward"),
                dict(step="all", label="All")
            ])
        ),
        rangeslider=dict(visible=True),  # adds a slider below chart
        type="date"
    )
)

#fig.show()
```

4.Give the complete timeline of Twitter in the stock
market (Line Graph):


```python
Twitter['Date'] = pd.to_datetime(Twitter['Date'], dayfirst=True, errors='coerce')
Twitter = Twitter.sort_values('Date')
Twitter = Twitter.dropna(subset=['Adj Close'])

# Plot Adjusted Close timeline
plt.figure(figsize=(14,6))
plt.plot(Twitter['Date'], Twitter['Adj Close'], label="Adj Close", color="blue", linewidth=1.5)

# Add labels and formatting
plt.title("Twitter Stock Price Timeline (2018 -> Present)", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Price (USD)", fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```
